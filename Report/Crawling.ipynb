```python
from newspaper import Article

# 크롤링할 뉴스 기사 URL
url = 'https://www.mk.co.kr/news/stock/11161449'

# 기사 객체 생성 및 다운로드
article = Article(url)
article.download()
article.parse()

# 기사 내용 추출
title = article.title
text = article.text
print(f"Title: {title}")
print(f"Content: {text}")
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# VADER 감정 분석기 초기화
sid = SentimentIntensityAnalyzer()

# 뉴스 기사 내용에 대한 감정 분석
sentiment = sid.polarity_scores(text)

# 감정 분석 결과 출력
if sentiment['compound'] >= 0.05:
    print("호재: 긍정적인 뉴스입니다.")
elif sentiment['compound'] <= -0.05:
    print("악재: 부정적인 뉴스입니다.")
else:
    print("중립: 감정이 긍정적이지도 부정적이지도 않습니다.")
```
