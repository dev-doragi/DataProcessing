```python
def get_news_links(keyword, base_url="https://www.mk.co.kr/news/ranking/stock/"):
    """
    뉴스 사이트에서 주어진 키워드가 포함된 뉴스 제목과 링크를 추출하는 함수
    """
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    # h3 태그를 찾고, 클래스가 'news_ttl'인 요소들만 필터링
    articles = soup.find_all("h3", class_="news_ttl")  # 뉴스 제목이 포함된 선택자

    news_links = []
    if not articles:
        print("뉴스 기사가 없습니다. 페이지 구조를 확인하세요.")
    else:
        for article in articles:
            # 제목 추출
            title = article.get_text(strip=True)
            
            # 링크 추출
            link_tag = article.find_parent("a")  # <a> 태그 부모 찾기
            if link_tag and link_tag.has_attr("href"):
                href = link_tag["href"]
                # 상대 경로를 절대 경로로 변환
                if href.startswith("/"):
                    link = "https://www.mk.co.kr" + href
                else:
                    link = href
            else:
                link = "링크 없음"
            
            # 키워드가 제목에 포함된 기사만 필터링
            if keyword in title and link != "링크 없음":
                news_links.append({"title": title, "link": link})
                print(f"제목: {title}")
                print(f"링크: {link}\n")
    
    print(f"크롤링된 기사 수: {len(news_links)}개\n")
    return news_links

def fetch_article_content(url):
    """
    주어진 뉴스 기사 URL에서 제목과 본문을 추출하는 함수
    """
    article = Article(url)
    article.download()
    article.parse()
    
    return article.title, article.text

def get_news_content_by_links(news_links):
    """
    키워드를 사용하여 뉴스 링크를 추출하고, 각 링크에서 기사 내용을 가져오는 함수.
    각 기사의 제목과 내용을 배열로 처리하여 반환.
    """
    news_data = []  # 제목과 내용을 저장할 리스트
    
    for news in news_links:

        title, content = fetch_article_content(news["link"])
        print(f"제목: {title}\n")
        print(f"내용: {content}\n")
        print("=======================================================================\
===============================================================================\n")
        news_data.append({"title": title, "content": content})  # 제목과 내용을 딕셔너리 형태로 리스트에 추가
    
    return news_data
```

```python
import requests
from bs4 import BeautifulSoup
from newspaper import Article

def get_top_news_links(base_url="https://www.mk.co.kr/news/ranking/stock/", num_articles=20):
    """
    뉴스 사이트에서 상위 n개의 뉴스 제목과 링크를 추출하는 함수.
    """
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    # h3 태그를 찾고, 클래스가 'news_ttl'인 요소들만 필터링
    articles = soup.find_all("h3", class_="news_ttl")  # 뉴스 제목이 포함된 선택자

    news_links = []
    if not articles:
        print("뉴스 기사가 없습니다. 페이지 구조를 확인하세요.")
    else:
        for article in articles[:num_articles]:  # 상위 n개 기사만 가져옴
            # 제목 추출
            title = article.get_text(strip=True)
            
            # 링크 추출
            link_tag = article.find_parent("a")  # <a> 태그 부모 찾기
            if link_tag and link_tag.has_attr("href"):
                href = link_tag["href"]
                # 상대 경로를 절대 경로로 변환
                if href.startswith("/"):
                    link = "https://www.mk.co.kr" + href
                else:
                    link = href
            else:
                link = "링크 없음"
            
            if link != "링크 없음":
                news_links.append({"title": title, "link": link})
                print(f"제목: {title}")
                print(f"링크: {link}\n")
    
    print(f"크롤링된 기사 수: {len(news_links)}개\n")
    return news_links

def fetch_article_content(url):
    """
    주어진 뉴스 기사 URL에서 제목과 본문을 추출하는 함수
    """
    article = Article(url)
    article.download()
    article.parse()
    
    return article.title, article.text

def get_news_content_by_links(news_links):
    """
    상위 n개의 뉴스 링크를 가져와, 각 링크에서 기사 내용을 추출하는 함수.
    각 기사의 제목과 내용을 배열로 처리하여 반환.
    """
    news_data = []  # 제목과 내용을 저장할 리스트
    
    for news in news_links:
        title, content = fetch_article_content(news["link"])
        print(f"제목: {title}\n")
        print(f"내용: {content}\n")
        print("=======================================================================\
===============================================================================\n")
        news_data.append({"title": title, "content": content})  # 제목과 내용을 딕셔너리 형태로 리스트에 추가
    
    return news_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 뉴스 내용이 딕셔너리로 되어 있다면, 'content'만 추출
documents = [news['content'] for news in news_content]

# TF-IDF 벡터화, 불용어 처리하고, 2-그램(n-gram)을 사용하여 2개 단어를 묶어서 분석
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), min_df=2)  # 불용어 제거, n-gram 1~2, 최소 빈도 2
X = vectorizer.fit_transform(documents)

# 키워드 추출
feature_names = vectorizer.get_feature_names_out()
for doc_idx, doc in enumerate(documents):
    print(f"문서 {doc_idx + 1} 키워드:")
    
    # 해당 문서의 TF-IDF 점수 배열을 가져옴
    tfidf_scores = X[doc_idx].toarray().flatten()  # .flatten()을 사용하여 1D 배열로 변환
    
    # 각 단어와 점수를 zip으로 묶고, 점수를 기준으로 내림차순 정렬
    scores = zip(feature_names, tfidf_scores)
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    
    # 상위 5개 키워드 출력
    for word, score in sorted_scores[:5]:  
        print(f"{word}: {score:.4f}")  # 점수는 소수점 4자리까지 출력
    print("\n")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

# 한국어 형태소 분석기 (Okt) 객체 생성
okt = Okt()

def extract_nouns(text):
    """
    텍스트에서 명사만 추출하는 함수
    """
    return okt.nouns(text)

def filter_proper_nouns(nouns):
    """
    명사에서 길이가 2 이상인 고유 명사만 필터링하는 함수
    """
    return ' '.join([noun for noun in nouns if len(noun) > 1])

def preprocess_documents(documents):
    """
    문서에서 명사를 추출하고 고유 명사만 필터링하는 함수
    """
    # 각 문서에서 고유 명사만 추출
    documents_nouns = [extract_nouns(doc) for doc in documents]
    # 길이가 2 이상인 고유 명사만 남기기
    documents_proper_nouns = [filter_proper_nouns(nouns) for nouns in documents_nouns]
    # 빈 문서 제외
    documents_proper_nouns = [doc for doc in documents_proper_nouns if doc.strip()]
    return documents_proper_nouns

def extract_keywords(documents_proper_nouns, ngram_range=(1, 2), min_df=2, top_n=5):
    """
    문서에서 TF-IDF를 사용하여 상위 n개의 키워드를 추출하는 함수
    """
    # TF-IDF 벡터화
    vectorizer = TfidfVectorizer(stop_words=None, ngram_range=ngram_range, min_df=min_df)
    X = vectorizer.fit_transform(documents_proper_nouns)

    # 키워드 추출
    feature_names = vectorizer.get_feature_names_out()
    keywords = []

    for doc_idx, doc in enumerate(documents_proper_nouns):
        doc_keywords = []
        # 해당 문서의 TF-IDF 점수 배열을 가져옴
        tfidf_scores = X[doc_idx].toarray().flatten()
        
        # 각 단어와 점수를 zip으로 묶고, 점수를 기준으로 내림차순 정렬
        scores = zip(feature_names, tfidf_scores)
        sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
        
        # 상위 n개 키워드 추출
        for word, score in sorted_scores[:top_n]:
            doc_keywords.append((word, score))
        
        keywords.append(doc_keywords)
    
    return keywords

def print_keywords(keywords):
    """
    추출된 키워드를 출력하는 함수
    """
    for doc_idx, doc_keywords in enumerate(keywords):
        print(f"문서 {doc_idx + 1} 키워드:")
        for word, score in doc_keywords:
            print(f"{word}: {score:.4f}")  # 점수는 소수점 4자리까지 출력
        print("\n")

# 예시 사용법
documents = [news['content'] for news in news_content]  # 뉴스 내용이 담긴 리스트

# 1. 문서 전처리 (명사 추출 및 고유 명사 필터링)
processed_documents = preprocess_documents(documents)

# 2. 키워드 추출
keywords = extract_keywords(processed_documents)
```

```python
def split_words(documents):
    """
    문서에서 각 단어를 하나씩 분리하여 저장하는 함수
    """
    # 각 문서를 공백을 기준으로 분리하여 단어 리스트로 변환
    return [doc.split() for doc in documents]

# 1. 문서를 단어별로 분리
words_list = split_words(processed_documents)

# 결과 출력
for idx, words in enumerate(words_list):
    print(f"문서 {idx + 1}의 단어들: {words[:10]}...")  # 각 문서의 앞 10개 단어만 예시로 출력
```

```python
from collections import Counter
from nltk import ngrams

# words_list에서 모든 단어를 합침
words = [word for doc in words_list for word in doc]

# 불용어 리스트
stopwords = set(['투자', '증권', '지난', '이후', '주가', '사진', '관계', '대비', '기록', '규모', '거래', '매수', '시장', '기준', '콤마', '시트', '맥락'])

# 불용어 제거
filtered_words = [word for word in words if word not in stopwords]

# 단일 단어와 2-gram을 각각 생성 및 빈도수 측정
single_word_counter = Counter(filtered_words)
bigrams = list(ngrams(filtered_words, 2))
bigram_strings = [' '.join(bigram) for bigram in bigrams]
bigram_counter = Counter(bigram_strings)

# 단일 단어와 2-gram에서 각각 상위 15개 선택
top_single_words = dict(single_word_counter.most_common(20))
top_bigrams = dict(bigram_counter.most_common(20))

# 결과 결합
combined_top_words = {**top_single_words, **top_bigrams}

# 출력
combined_top_words
```
