```python
#  재무 데이터 기준 종목투자 시뮬레이션

folder_dir = 'C:/data\실습데이터'
DATA_FILE = "Kospi_Fin_Price_2024.xlsx"

data_wb = pd.ExcelFile(folder_dir + "/" + DATA_FILE)

find_ts = data_wb.parse("재무", index_col=0)  #  재무데이터 : 인덱스는 회계 연도

find_ts.head(10)

adj_close_ts = data_wb.parse("수정주가", index_col=0)  # 주가 데이터 : 인덱스는 date

# adj_close_ts.head(5)
```

```python
# 재무비율 만들기  :  재무데이터  /재무 데터터

find_ts ['ROE']=find_ts ['당기순이익']/find_ts["총자본"]
find_ts ['ROA']=find_ts ['당기순이익']/find_ts["총자산"]
find_ts ['부채비율']=find_ts ['총부채']/find_ts["총자본"]
find_ts ['매출액증가율']=find_ts ['매출액'].pct_change(1)
```

```python
# find_ts[['매출액','영업이익','당기순이익', 매출액증가율, 부채비율,'EPS','BPS','SPS','CFPS','ROE','ROA','부채비율']
```

```python
ps_ts= find_ts[['종목명','매출액증가율','ROE','ROA','부채비율']]  # 필요항목만 추출
```

```python
ps_ts.head(5)
```

```python
ps_ts = ps_ts[~ps_ts['종목명'].isin(['엘브이엠씨홀딩스','동아타이어','프레스티지바이오파마']) ]   # 최근 상장폐지 및 기타사유
```

```python
# 재무비율 순위에 따른 포트폴리오 구성

rebalancing_list = ['2011-04', '2012-04', '2013-04', '2014-04', '2015-04' ,
                    '2016-04', '2017-04','2018-04', '2019-04', '2020-04',   
                    '2021-04', '2022-04', '2023-04','2024-04']

fin_list=list(range(2010, 2024))

column_list = ['재무안정','재무불안정']  # 포트폴리오 일간수익률  : 성장성 vs 수익성

portfolio_ret_df = pd.DataFrame(columns=column_list, index = adj_close_ts.loc[rebalancing_list[0]:].index)

```

```python
# 매년 종목군을 변화시키면서 리밸런싱 한다.

for i in range(len(rebalancing_list)):
                  
    fin_combi_A =ps_ts.loc[fin_list[i]].query(' 부채비율<1').set_index('종목명')  
    fin_index_A_list =  fin_combi_A.index
    
    fin_combi_B =ps_ts.loc[fin_list[i]].query('부채비율>4').set_index('종목명')  
    fin_index_B_list =  fin_combi_B.index
       
    portfolio_ret_df['재무안정'].loc[rebalancing_list[i]:] = adj_close_ts[fin_index_A_list].loc[rebalancing_list[i]:].pct_change(1).mean(axis=1).dropna()
    portfolio_ret_df['재무불안정'].loc[rebalancing_list[i]:] = adj_close_ts[fin_index_B_list].loc[rebalancing_list[i]:].pct_change(1).mean(axis=1).dropna()    
        
   # print (fin_index_B_list)
```

```python
portfolio_ret_df.head(5)
```

```python
# 성과계산 함수

def index_100(ret_df):   # 누적성과 계산함수
    index_df = (1 + ret_df).cumprod() * 100
    return index_df

def ret_risk_profile(mp_index):  # 각종 성과지표 계산 함수
    ret_df = pd.DataFrame(index=mp_index.columns)
    ret_df["누적수익률"] = mp_index.iloc[-1] / mp_index.iloc[0] - 1
    delta_to_years = (mp_index.index[-1] - mp_index.index[0]).days / 365.25
    ret_df["누적수익률(연율화)"] = (1 + ret_df["누적수익률"]) ** (1 / delta_to_years) - 1
    ret_df["변동성(연율화)"] = mp_index.resample("W").last().pct_change().std() * np.sqrt(52)
    ret_df = ret_df * 100
    ret_df["샤프비율"] = ret_df["누적수익률(연율화)"] / ret_df["변동성(연율화)"]
    index_year = mp_index.resample("A").last()
    index_year_pct = index_year.pct_change()
    index_year_pct.iloc[0] = index_year.iloc[0] / 100 - 1
    index_year_pct.index = index_year_pct.index.year
    index_year_pct = index_year_pct * 100
    
    ret_df["음의 변동성(연율화)"] = mp_index.resample("W").last().pct_change() [mp_index.resample("W").last().pct_change()<0 ].std() * np.sqrt(52)
    ret_df["Sortino"] = ret_df["누적수익률(연율화)"] / ret_df["음의 변동성(연율화)"]
    
    comp_ret = (mp_index.pct_change()+1).cumprod() 
    peak=comp_ret.expanding(min_periods=1).max() 
    dd = (comp_ret/peak)-1 
    ret_df["Maxdrawdown"] =dd.min() 

    
    return pd.concat([ret_df.T, index_year_pct])
```

```python
kindex= adj_close_ts['kospi']['2011-04-01': ].pct_change(1)   # 주가지수 수익률
kindex.iloc[0]=0

portfolio_ret_df.iloc[0] = 0          # 포트폴리오 수익률
portfolio_ret_df = pd.concat([kindex,portfolio_ret_df],axis=1).dropna()

plot_list = ['kospi','재무안정','재무불안정']
index_100(portfolio_ret_df[plot_list]).plot()  # 누적성과 그래프

pp = ret_risk_profile(index_100(portfolio_ret_df[plot_list]))  # 성과 지표 계산

index_year=index_100(portfolio_ret_df[plot_list]).resample("A").last()  # 기준연도(2011) 대비 연간기준 성과
print(pp.round(2))  # 성과 지표 출력
```

##### 베타시뮬레이션 

```python
adj_all=adj_close_ts[adj_close_ts.index>'2010-01-23']  # 시작일자 지정
```

```python
# 베타값 구하기 위한 작업 (일간수익률)

adj_rate = adj_all.pct_change(1)   # 일간수익률 계산
adjr=adj_rate[1:]
adjr=adjr.dropna(axis='columns')   #  데이터가  기간내에 처음과  끝시점 모두 있는 종목만 추린다. 
adjr.head(5)
```

```python
# 베타값 구하기 위한 작업 (주간수익률)

adj_w = adj_all.resample('w').last()   #  주간 시계열로 변환
adj_wrate=adj_w.pct_change(1)         # 주간 수익률 변환
adj_wr=adj_wrate[1:]
adj_wr=adj_wr.dropna(axis='columns')   #  데이터가  기간내에 처음과  끝시점 모두 있는 종목만 추린다. 
adj_wr.head(5)
```

```python
# 베타구하는 함수

def beta(df, market=None):
    
    # if market is None:
    market = df['kospi']             # 독립변수 설정
    df = df.drop('kospi', axis=1)    # 종속변수 설정 (독립변수제외)
    X = market.values.reshape(-1, 1)   # 재구조화 (-1,1) 배열 : 열차원(1)에 따라 행의 갯수가변적 
    X = np.concatenate([np.ones_like(X), X], axis=1)
    b = np.linalg.pinv(X.T.dot(X)).dot(X.T).dot(df.values)
    return pd.Series(b[1], df.columns, name=df.index[-1])
```

```python
print(beta(adj_wr).sort_values)
```

```python
# 베타순위에 따라 상위 종목 추출 

beta_rank=beta(adj_wr).sort_values(ascending=True).head(10)
beta_rank.plot.bar(figsize=(7,2))
```

#### 롤링베타값을 구해 종목별 베타값 추이를 살펴본다,
롤링기간을 설정하고 연속해서 베타값을 구해간다.


```python
def roll(df, w):
    for i in range(df.shape[0] - w + 1):   # 총기간, 롤링 기간 설정
            yield pd.DataFrame(df.values[i:i+w, :], df.index[i:i+w], df.columns)      
```

```python
betas = pd.concat([beta(adj_wr) for adj_wr in roll(adj_wr, 52)], axis=1).T   # 주간: 52, 일간 240
```
